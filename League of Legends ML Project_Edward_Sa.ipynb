```python
%matplotlib inline
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import random


# read the csv file using pandas
data = pd.read_csv("data_summary_cleansed.csv", sep=",", header = 0)
columns = []
# add columns from csv into an array
for column in data:
    columns.append(column)
# now we have a list of arrays (each array is separate champion with associated values), use this for algorithms
dataset = np.array(data)
# should be an array of arrays
print(dataset)
```

```python
# Display data in a nice table
data
```

```python
# Linear regression on kills and damage dealt to champions (physical + magic)
from sklearn import linear_model
# need to get two arrays, one for kill values and one for damage values
kills = []
damage = []
# go through every array (champion) in data set and grab correct data and add to corresponding array
for data in dataset:
    kills.append([data[2]])
    damage.append([data[7] + data[8]])

# length of kills/damage is 126 (there's 126 champions)
x_train = kills[0:100]
y_train = damage[0:100]
x_test = kills[100:]
linear = linear_model.LinearRegression()
linear.fit(x_train, y_train)
linear.score(x_train, y_train)
print('Coefficient: \n', linear.coef_)
print('Intercept: \n', linear.intercept_)
predicted= linear.predict(x_train)

plt.scatter(x = x_train, y = y_train)
plt.plot(x_train,predicted)
print('Error: \n', sum((predicted-(y_train))**2))
```

Used regression to see if there was a high correlation/definitive linear formula to determine total damage dealt to champions based on number of kills. The result showed not too strong of a correlation, which matches with how I predicted. This is because in League of Legends the person who gets the kill is who deals the finishing blow, it is not percent damage based.

```python
pair = []
# for every element in SVM array will be an array with 2 values: KDA and total damage dealt
# each element in the array is a champion
# will only be doing 25 champions, as classifying all 126's role will take too long since info not given
for champ in dataset[:25]:
    KDA = (champ[2] + champ[4])/champ[3]
    total_dmg = champ[7] + champ[8]
    pair.append([KDA, total_dmg])

# need to classify each champion's target value: 
# 0 =  TOP, 1 = JUNGLE, 2 = MID, 3 = ADC, 4 = SUPPORT
# classified first 25 champions manually
target = [0,2,2,4,1,2,2,3,2,4,4,2,4,3,2,0,2,0,2,3,0,2,1,1,3]
targett = np.asarray(target)
x = []
y = []
for elem in pair:
    x.append(elem[0])
    y.append(elem[1])
```

SVM to classify types of champions. Real result would be five categories
(League of Legends has five roles: top, jungle, middle, adc, support)

Pair values to use will be KDA (Kill Death Assist Ratio) and total damage dealt to champions.

```python
from sklearn import svm, datasets
import matplotlib as mp
import matplotlib.pyplot as plt

# Plot resulting Support Vector boundaries with original data
# Create fake input data for prediction that we will use for plotting
# This code is based off of one given in lecture, but instead of iris dataset we are using League of Legends champions
pair_min, pair_max = min(x) - 1, max(x) + 1
target_min, target_max = min(y) - 1, max(y) + 1
h = (pair_max / pair_min)/100
xx, yy = np.meshgrid(np.arange(pair_min, pair_max, h),
 np.arange(target_min, target_max, h))
X_plot = np.c_[xx.ravel(), yy.ravel()]

# Create the SVC model object
C = 1.0 # SVM regularization parameter
svc = svm.SVC(kernel='linear', C=C, decision_function_shape='ovr').fit(pair, target)
Z = svc.predict(X_plot)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.contourf(xx, yy, Z, cmap=plt.cm.tab10, alpha=0.3)
plt.scatter(x, y, c=targett, cmap=plt.cm.Set1)
plt.xlabel('KDA')
plt.ylabel('Total Damage Dealt')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with Linear Kernel')
```

With having 5 types of champions, the result of the SVC shows that KDA does not have too much of a role in guessing what position the champion is played as. Ths is because the regions are only separated into different vertical regions, which is only affected by the y axis (Total Damage Dealt). This result goes along well with the logic behind League of Legends gameplay, because typically those in supportive roles will deal less damage compared to those in high carry roles. The KDA not playing too much of a role goes along with this logic too, because we are taking the average of all player data, it's important to note that although a supportive role may take in more assists and less kills than a carry role, since KDA is computed by (Kills + Assists)/Deaths, so there is no differentiating between kills and assists.
